```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%pylab inline
```

```python
#importing the csv file
trr_base_df = pd.read_csv('./files/Terrorism_Case_2001_to_today/people_protected.csv')
#trr_base_df.head()
```

```python
#Calculating total Men and Women based on gender, involved in Terrorism and plotting a pie chart

trr_men_women = trr_base_df['person_ID'].groupby(trr_base_df['gender']).count().tolist()

# Data to pie chart based on the above data
labels = 'Female', 'Male'
colors = ['lightskyblue', 'red']
explode = (0.1, 0)
 
# Plot
plt.pie(trr_men_women,explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140,colors=colors)
 
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#Calculating the number of convicted, based on their citizenship

trr_temp = trr_base_df.groupby('citizenship_status').count()
trr_temp = trr_temp.reset_index()
trr_citizenship = trr_temp[['citizenship_status','person_ID']]
trr_citizenship = trr_citizenship.rename(columns={'person_ID':'total_count'})
citizenship_status = trr_citizenship['citizenship_status'].tolist()
total_count = trr_citizenship['total_count'].tolist()

#Draw a bar graph based on the above data
x = range(len(citizenship_status))
y = total_count
print y
width = 1/1.5
plt.bar(x, y, width, color="blue")
plt.show()
```
